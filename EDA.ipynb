```python
country_code = pd.read_excel("Country-Code.xlsx")
```

```python
dataset_final = pd.merge(dataset, country_code, on="Country Code", how="left")
```

```python
country_names = dataset_final.Country.value_counts().index
```

```python
country_values = dataset_final.Country.value_counts().values
```

```python
country_names
```

```python
plt.pie(country_values[0:3], labels=country_names[0:3])
```
