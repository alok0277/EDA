```python
# Using train test split to spliting the dataset into sample testing and training the model
from sklearn.model_selection import train_test_split
```

```python
y=dataset["Aggregate rating"]
x=dataset.drop(["Aggregate rating"],axis=1)
print("Shape of y:",y.shape)
print("Shape of x:",x.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

```python
x_prop = len(X_train) / (len(X_train) + len(X_test))
xt_prop = len(X_test) / (len(X_train) + len(X_test))
plt.bar(["x-train","x_test"], [x_prop, xt_prop])
plt.show()
```
